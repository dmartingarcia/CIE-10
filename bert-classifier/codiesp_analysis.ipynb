```python
# TRAIN
##################

import pandas as pd
from collections import Counter

# Read the CSV file
df = pd.read_csv('codiesp_csvs/codiesp_D_source_train.csv')

# Convert string representation of list to actual list
df['labels'] = df['labels'].apply(eval)

# Flatten the list of labels and count frequencies
code_counts = pd.Series([code.strip().strip("'") for codes in df['labels'] for code in codes if code.strip()]).value_counts()

# Display the results
print("Code frequencies:")
print(code_counts)

child_counts = Counter([code for codes in df['labels'] for code in codes if '.' in code])
print("Top 10 hijos:", child_counts.most_common(10))
print("Clases con 1 ejemplo:", sum(1 for v in child_counts.values() if v == 1))

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.hist(code_counts.values, bins=50, log=True)
plt.xlabel('Frequency of occurrence')
plt.ylabel('Number of codes (log scale)')
plt.title('Distribution of Code Frequencies')
plt.grid(True, alpha=0.3)
plt.show()

# Get parent classes (codes before the dot)
parent_codes = pd.Series([code.split('.')[0] for codes in df['labels'] for code in codes]).value_counts()

# Create bar plot
plt.figure(figsize=(15, 8))
plt.bar(range(len(parent_codes)), parent_codes.values)
plt.ylabel('Frequency')
plt.title('Distribution of Parent Codes')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Top 10 most frequent codes
print("Top 10 most frequent codes:")
print(code_counts.head(10))

print("\nBottom 10 least frequent codes:")
print(code_counts.tail(10))

```
